```python
# Set Axis and drop diagnosis
x=data.drop(columns=["diagnosis"])
y=data["diagnosis"]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
data.head(5)
```

```python
# Split dataset into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,shuffle=True)
```

```python
from sklearn.neighbors import KNeighborsClassifier

KNN=KNeighborsClassifier()
KNN.fit(x_train,y_train)
y_prediction=KNN.predict(x_test)

print("Score of x_test, y_test",KNN.score(x_test,y_test))
print("Score of x_train, y_trian",KNN.score(x_train,y_train))
```

```python
#Create Confusion Matrix
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test,y_prediction)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Benign', 'Malignant'])
disp.plot()


plt.title("Confusion Matrix of KNN")
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_prediction))
```

```python
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier()

random_forest.fit(x_train, y_train)
labels=random_forest.predict(x_test)
score=random_forest.score(x_test, y_test)

print("Score of Random Forest:", score)
```

```python
#Tuning 
from sklearn.model_selection import GridSearchCV


parameters={
    "n_estimators": [100,200,500],
    #"max_features": ["auto","sqrt", "log2"],
    #"criterion" : ["gini", "entropy"],
    "max_depth": [5, 10, 15],
    "min_samples_split": [2, 5, 10]
}

grid_search= GridSearchCV(
    estimator= random_forest,
    param_grid= parameters,
    scoring= "accuracy",
    cv= 5
)

grid_search= grid_search.fit(x,y)

print(f"Best Parameters for Random Forest: {grid_search.best_params_}Accuracy of Random Forest Algorithm:{grid_search.best_score_} % ")
```
